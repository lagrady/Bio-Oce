```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
#Make the dataframe
df1 = pd.read_csv("CD_Leg1.csv", delimiter=",")
df1
```

```python
df2 = pd.read_csv("CD_Leg2.csv", delimiter=",")
df2
#df2_sub = df2.dropna()
#df2_sub
```

```python
#Plotting Cruise 1
#Temperature with depth
plt.figure(figsize = (10,15))
plt.subplot(221)
plt.plot(df1.iloc[:,8], -df1.iloc[:,4], '.r')
plt.ylabel('Depth (m)')
plt.xlabel('Temperature (Celsius)')
plt.title('Temperature with Depth (Cruise 1)')

#Salinity with depth
plt.subplot(222)
plt.plot(df1.iloc[:,2], -df1.iloc[:,4], '-b')
plt.ylabel('Depth (m)')
plt.xlabel('Salinity (PSU)')
plt.title('Salinity with Depth (Cruise 1)')

#Oxygen with depth
plt.subplot(223)
plt.plot(df1.iloc[:,6], -df1.iloc[:,4], '-g')
plt.ylabel('Depth (m)')
plt.xlabel('Oxygen (%)')
plt.title('Oxygen with Depth (Cruise 1)')

#Light Attenuation with depth
plt.subplot(224)
plt.plot(df1.iloc[:,10], -df1.iloc[:,4], '-y')
plt.axvline(x=.01, c='k')
plt.ylabel('Depth (m)')
plt.xlabel('Light Attenuation (%)')
plt.title('Light Attenuation with Depth (Cruise 1)')
```
